```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
import matplotlib.pyplot as plt

# Dropping the duplicate "fips " column
df = df.drop(columns=['fips '])

# Sorting the data by state and year
df = df.sort_values(['name', 'year'])

# Calculating the year-over-year income growth rate
df['YoY Income Growth Rate'] = df.groupby('name')['income'].pct_change() * 100

# Plotting the results
plt.figure(figsize=(15,10))

for name, group in df.groupby('name'):
    plt.plot(group['year'], group['YoY Income Growth Rate'], label=name)

plt.xlabel('Year')
plt.ylabel('Year-over-Year Income Growth Rate (%)')
plt.title('Year-over-Year Income Growth Rate for Each State')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming the DataFrame is named df
df = df[['fips', 'income', 'year', 'name']].copy()

# Calculate the year over year income growth rate
df['income_growth'] = df.groupby('name')['income'].pct_change() * 100

# Create a pivot table for better visualization
pivot_df = df.pivot(index='year', columns='name', values='income_growth')

# Plotting the data
plt.figure(figsize=(15,10))
for column in pivot_df.columns.to_list():
    plt.plot(pivot_df.index, pivot_df[column], marker='', linewidth=1, alpha=0.9, label=column)

plt.legend(loc=2, ncol=2)
plt.title("Year over Year Income Growth Rate for Different States", loc='left', fontsize=12, fontweight=0, color='orange')
plt.xlabel("Year")
plt.ylabel("Growth Rate (%)")
plt.show()
```
